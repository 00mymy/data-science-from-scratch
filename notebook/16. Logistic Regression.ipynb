### Linear Regression을 이용한 Paid/Unpaid Classification
* 예측값 범위에 따라 대략 구분이 되기는 함
  + 예측값 < 0: Unpaid (실제값=0)
  + 0 <= 예측값 < 0.7: Paid/Unpaid가 섞여 있음
  + 0.7 < 예측값: Paid (실제값=1)
* 대략 0~0.7 구간의 어느 값을 기준으로 Paid/Unpaid를 판단할 수 있을 것 같다
  + 0~0.7 구간의 값 중 하나를 택하여 Paid/Unpaid를 판단할 수도 있겠지만, 좀 더 견고한 해답을 찾아보자
* 무식한 방법
  + 예) Linear Regression 결과를 기준으로, Paid/Unpaid 경계값을 변경시켜가면서 가장 Accuracy가 높은 경계값을 찾아보자

```python
def paid_or_not(y_hat_value, cutoff = 0.25):
    return y_hat_value > cutoff

def test_accuracy(y_hat, y, cutoff):
    num_correct = 0
    total = len(data)
    for i in range(total):
        if paid_or_not(y_hat[i], cutoff) == y[i]:
            num_correct += 1
  
    return num_correct / len(data)
```

```python
import numpy as np

t_x = []
t_y = []
for s in np.arange(min(predictions),max(predictions), 0.01):
    t_x.append(s)
    t_y.append(test_accuracy(predictions, ys, s))

plt.plot(t_x,t_y)
```

# The Logistic Function

### Linear Regression - 뭔가 부족하다
* Paid/Unpaid 실제값은 0 또는 1인데, Linear Regression 예측값은 0보다 작은값부터 1보다 큰 값가지 다양하게 나옴
  + 회귀계수($\beta_i$) 값에 따라 0 보다 훨씬 작은값이나 1보다 훨씬 큰 값이 나올 수도 있다
    - 실제값과 예측값의 차이(error)가 크게 달라질 수 있다.
  + 예측 결과는 0 또는 1이 나와야한다 (Yes/No Classifier)
  + 앞에서 한 것처럼 판단 기준이 되는 값을 모델 만들 때마다 측정하면서 판단하는 것은 불편하다
  + 모델이 예측하는 값이 크면 클수록 1이 나올 가능성이 높고, 작아질수록 0이 나올 가능성이 높아지면 좋겠다
    - 확률적인 의미가 있으면 좋다
    - 예) 예측값 0.25 &rarr; Yes일 확률이 25%

### Logistic 함수를 사용하면 된다
* 시그모이드(Sigmoid) 함수의 하나
  + S자 모양의 함수
  + $f(x)={1\over{(1+\exp^{-x})}}$
  + 중간영역에서는 Linear하게 증감하지만, 큰 값이나 작은 값으로 갈수록 특정 값으로 수렴하는 특징을 가진다
    - 모델이 예측하는 값이 크면 클수록 1이 나올 가능성이 높고, 작아질수록 0이 나올 가능성이 높아지도록 변환시킬 수 있다
  + Wiki: [시그모이드 함수](https://ko.wikipedia.org/wiki/%EC%8B%9C%EA%B7%B8%EB%AA%A8%EC%9D%B4%EB%93%9C_%ED%95%A8%EC%88%98)
* Linear Regression 결과에 Logistic 함수 적용
* $y_i=f(x_i\beta) + \epsilon_i$

```python
import math

def logistic(x: float) -> float:
    return 1.0 / (1 + math.exp(-x))
```

### Likelihood Function
* Linear Regression에서는 SSE(Sum of Squared Errors)를 최소화하는 $\beta$를 구했음
* Logistic Regression에서는 "Maximum-likelihood estimation" 사용
  + Error를 최소화하는 방식으로 풀기 어려운 문제에 적용
* Likelihood Function
  + $p(y_i|x_i,\beta) = f(x_i\beta)^{y_1}(1-f(x_i\beta)^{1-y_1}$
  + 주어진 입력 $x_i$와 $\beta$에 대해, Logistic 함수 $f$는 다음과 같은 Likelihood를 제공한다
    - $y_i$가 1일 Likelihood = $f(x_i\beta)$
    - $y_i$가 0일 Likelihood = $1-f(x_i\beta)$
    - Likelihood $\approx$ 확률
  + $\log(p(y_i|x_i,\beta)) = f(x_i\beta)^{y_1}(1-f(x_i\beta)^{1-y_1}$

### Gradient Descent 적용하기
* Gradient Descent를 적용하기 위하여 Cost Funtion과 이의 미분함수가 필요
  + Log-Likelihood 함수
    - $LL(\beta|x_i,y_i)=y_i\log f(x_i\beta) + (1-y_i)\log(1-f(x_i\beta))$
  + Log-Likelihood 함수의 미분함수
    - $(y_i - f(x_i\beta))x_i$
* minimize 대신 maximize
  + Error를 최소화하는 대신 Likelihood를 최대화하는 $\beta$를 구해야 한다
* Negative Log-Likelihood
  + Log-Likelihood **최대화** $\approx$ Negative Log-Likelihood **최소화**
  + Gradient Descent 알고리즘을 이용하여 Negative Log-Likelihood를 최소화하는 $\beta$를 구하면 된다
* Overall Likelihood
  + 개별 Likelihood의 **곱**
  + Overall Log-Likelihood: 개별 Log-Likelihood의 **합** (Log의 성질)

```python
from scratch.linear_algebra import Vector, dot

def _negative_log_likelihood(x: Vector, y: float, beta: Vector) -> float:
    """The negative log likelihood for one data point"""
    if y == 1:
        return -math.log(logistic(dot(x, beta)))
    else:
        return -math.log(1 - logistic(dot(x, beta)))
```

```python
from typing import List

def negative_log_likelihood(xs: List[Vector],
                            ys: List[float],
                            beta: Vector) -> float:
    return sum(_negative_log_likelihood(x, y, beta)
               for x, y in zip(xs, ys))
```

```python
from scratch.linear_algebra import vector_sum

def _negative_log_partial_j(x: Vector, y: float, beta: Vector, j: int) -> float:
    """
    The j-th partial derivative for one data pont
    here i is the index of the data point
    """
    return -(y - logistic(dot(x, beta))) * x[j]

def _negative_log_gradient(x: Vector, y: float, beta: Vector) -> Vector:
    """
    The gradient for one data point
    """
    return [_negative_log_partial_j(x, y, beta, j)
            for j in range(len(beta))]

def negative_log_gradient(xs: List[Vector],
                          ys: List[float],
                          beta: Vector) -> Vector:
    return vector_sum([_negative_log_gradient(x, y, beta)
                       for x, y in zip(xs, ys)])
```

### Classification
* 0과 1사이의 값 &rarr; 0 또는 1 (Categorical)
* Decision boundary: 0.5
  + Logistic Function은 입력 값에 대해 0~1 사이의 확률을 제공하므로 0.5를 기준으로 판단하면 된다 (대충 0.5 찍은 것이 아님)
  + 의도적으로 예측 수준을 높이거나 낮추기 위해 Decision boundary를 조정할 수도 있다
    - 예) 암 진단: 오류 가능성이 높더라도 어지간하면 암이라고 판단하는게 안전하다 (Decision boundary < 0.5)
    - 예) 암 완치 판정: 상당히 높은 수준의 확신(높은 확률)이 들 때 완치라고 판정해야 한다  (Decision boundary > 0.5)
* Precision과 Recall 측정, 시각화를 통해 예측 성능을 검증해 볼 수 있다

```python
true_positives = false_positives = true_negatives = false_negatives = 0

for x_i, y_i in zip(x_test, y_test):
    prediction = logistic(dot(beta, x_i))

    if y_i == 1 and prediction >= 0.5:  # TP: paid and we predict paid
        true_positives += 1
    elif y_i == 1:                      # FN: paid and we predict unpaid
        false_negatives += 1
    elif prediction >= 0.5:             # FP: unpaid and we predict paid
        false_positives += 1
    else:                               # TN: unpaid and we predict unpaid
        true_negatives += 1

precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)



print(precision, recall)

assert precision == 0.75
assert recall == 0.8
```

### Predict
* 교재에 제시된 코드를 좀 더 정리해 보자
* predictor 정의
* "Chap 11. Machine Learning"에서 정의한 측정 함수를 사용하여 Accuracy, Precision, Recall, F1-Score를 측정해 보자

```python
def predict(beta, rescaled_x, threshould=0.5):
    prediction = logistic(dot(beta, rescaled_x))

    if prediction >= 0.5:
        return 1 #Yes
    
    return 0 # No
```

```python
from collections import defaultdict
def test_cm(beta, x_test, y_test):
    comfustion_matrix = defaultdict(int)
    for x_i, y_i in zip(x_test, y_test): 
        prediction = predict(beta, x_i)

        if y_i == 1 and prediction == 1:  # TP: paid and we predict paid
            comfustion_matrix['tp'] += 1
        elif y_i == 1:                    # FN: paid and we predict unpaid
            comfustion_matrix['fn'] += 1
        elif prediction == 1:             # FP: unpaid and we predict paid
            comfustion_matrix['fp'] += 1
        else:                             # TN: unpaid and we predict unpaid
            comfustion_matrix['tn'] += 1
    return comfustion_matrix

from scratch.machine_learning import accuracy, precision, recall, f1_score
cm = test_cm(beta, x_test, y_test)
print( 'Accuracy ', accuracy(cm['tp'], cm['fp'], cm['fn'], cm['tn']))
print( 'Precision', precision(cm['tp'], cm['fp'], cm['fn'], cm['tn']))
print( 'Recall   ', recall(cm['tp'], cm['fp'], cm['fn'], cm['tn']))
print( 'F1-Score ', f1_score(cm['tp'], cm['fp'], cm['fn'], cm['tn']))
```
